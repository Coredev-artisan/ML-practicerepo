### REAL-WORLD PROBLEMS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 120000, 180000, 250000, 230000],
    [2, 120000, 140000,160000,180000],
    [3, 150000, 230000,170000,190000],
    [4, 120000, 150000, 240000,130000],
    [5,140000, 170000,140000, 240000]
])
print("zomato =====sale_data=====is")
```

```python
print("\n sale data shape", sales_data.shape)
```

```python
print("\n the data of 1st 3 restaurant is: ", sales_data[0:3])
```

```python
print(np.sum(sales_data[:,1:],axis =0))
```

```python
print(np.min(sales_data[:,1:],axis=0))
```

```python
print(np.max(sales_data[:, 1:],axis=1))
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis=0)
print(np.cumsum(sales_data[:,1:],axis=0))
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Avarage cumsum")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
